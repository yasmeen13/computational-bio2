```python
import networkx as nx
```

```python
G =nx.Graph()
G.add_edges_from([('c','a'),('a','d'),('d','b'),('b','e'),('e','c')])
```

```python
nx.draw(G,with_labels=True,node_color="pink",edge_color="green")
```

```python
dict(G.degree())
```

```python
nx.to_numpy_matrix(G)
```

```python
df=pd.read_csv('SaccharomycesCerevisiaeS288C_binary_hq.txt',nrows=100,sep='\t',usecols=['Uniprot_A','Uniprot_B'])
```

```python
bio_graph=nx.from_pandas_edgelist(df,'Uniprot_A','Uniprot_B',create_using=nx.DiGraph())
```

```python
nx.draw(bio_graph,with_labels=True)
```

```python
d=dict(bio_graph.degree())
```

```python
sorted(d.items(),key=lambda t:t[1])
```

```python
import matplotlib.pyplot as plt
```

```python
order=sorted(d.items(),key=lambda t:t[1])
x,y=zip(*order)
```

```python
plt.plot(x,y)
```
